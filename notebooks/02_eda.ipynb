# Exploratory data analysis

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## load the dataset

```python

df = pd.read_csv("../data/raw/train.csv")
df.head()
```

## basic database information

```python
# Shape of dataset
print("Shape:", df.shape)

# Info about columns (data types, nulls)
df.info()

# Summary statistics
df.describe()
```

## missing values analysis

```python
# Count missing values per column
missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]
print("Missing values:\n", missing)

# Visualize missing values
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

## Target variable definition

```python
plt.figure(figsize=(10,5))
sns.histplot(df["SalePrice"], kde=True, bins=40)
plt.title("Distribution of Sale Price")
plt.show()
```

## corelation with target

```python
corr_with_target = df.select_dtypes(include=['number']).corr()['SalePrice'].sort_values(ascending=False)

plt.figure(figsize=(12,8))
sns.heatmap(corr_with_target.to_frame(), annot=True, cmap="coolwarm")
plt.title("Correlation of Features with SalePrice")
plt.show()

```

## Numerical feature analysis

```python
num_cols = df.select_dtypes(include=[np.number]).columns

# Example: GrLivArea vs SalePrice
plt.figure(figsize=(10,6))
sns.scatterplot(x="GrLivArea", y="SalePrice", data=df)
plt.title("Living Area vs Sale Price")
plt.show()
```

## categorial feature analysis

```python
cat_cols = df.select_dtypes(include=[object]).columns

# Example: Neighborhood vs SalePrice
plt.figure(figsize=(12,6))
sns.boxplot(x="Neighborhood", y="SalePrice", data=df)
plt.xticks(rotation=45)
plt.title("Neighborhood vs Sale Price")
plt.show()
```
